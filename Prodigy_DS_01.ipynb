```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_2631256.csv'
data = pd.read_csv(csv_file_path, skiprows=4)
data.head()
```

```python
selected_countries = ['Aruba', 'Afghanistan', 'Angola', 'Australia', 'Brazil', 'Canada', 'China', 'India', 'United States']
subset = data[data['Country Name'].isin(selected_countries)]
```

```python
population_2023 = subset[['Country Name', '2023']]
```

```python
plt.figure(figsize=(12, 6))
plt.bar(population_2023['Country Name'], population_2023['2023'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Population of Selected Countries in 2023')
plt.xticks(rotation=45)
plt.show()
```
